```python
def load_json(path):
    with open(path) as f:
        data = [json.loads(line) for line in f]
    return data

def load_json_file(path):
    with open(path, 'r', encoding='utf8') as f:
        return json.load(f)

def create_dir(dir_path):
    if not os.path.exists(dir_path):
        os.makedirs(dir_path)
```
