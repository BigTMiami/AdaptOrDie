# Summary
This takes a 500k review dataset, compiles the model,  saves checkpoints every 100k reviews and pushes the final model to the hub.

# Results


# Details
* Uses batch size of 34 and gradient accumulation of 10
  * Gradient Updates every 340 rows (34 * 10)
* Uses 1 epoch
* Dataset
  * 83,900 rows - train
    * 246 update steps using batch size of 34 * accumulation of 10
  * 8,340 rows - validation
* Saves checkpoint every 50 steps
  * Only has one current save on the google drive
  * Saves checkpoint to hub on checkpoint save
  * Saves model at end as well
* Compile creates warning but it seems ok: popen_fork.py:66: RuntimeWarning: os.fork()

```python
from pynvml import *

def print_gpu_utilization():
    nvmlInit()
    handle = nvmlDeviceGetHandleByIndex(0)
    info = nvmlDeviceGetMemoryInfo(handle)
    gpu_used = info.used//1024**2
    print(f"GPU {gpu_used} MB")
```

```python
from time import time
```

```python
from datasets import load_dataset
dataset = load_dataset("BigTMiami/amazon_25M_500_000_condensed")
print(dataset)
print_gpu_utilization()
```

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("roberta-base")
block_size = tokenizer.model_max_length
print(f"block_size:{block_size}")
print_gpu_utilization()
```

```python
from transformers import AutoConfig, AutoModelForMaskedLM

config = AutoConfig.from_pretrained("roberta-base")
model = AutoModelForMaskedLM.from_config(config)
print_gpu_utilization()
```

# Train

```python
from transformers import Trainer, TrainingArguments

training_args = TrainingArguments(
    output_dir="amazon_pretraining_500k_reviews",
    learning_rate=0.0005, # Paper for DAPT training
    per_device_train_batch_size=34, # 346,m - TRYING SMALLER BATCH SIZE
    per_device_eval_batch_size=34, #346,m - TRYING SMALLER BATCH SIZE
    gradient_accumulation_steps=10, # Creates effective batch of 34*10*6
    num_train_epochs=1, # Just one epoch for this test, already compiled
    weight_decay=0.01,
    warmup_ratio=0.06, # Paper: warmup proportion of 0.06
    adam_epsilon=1e-6, # Paper 1e-6 (huggingface default 1e-08)
    adam_beta1=0.9, # Paper: Adam weights 0.9
    adam_beta2=0.98, # Paper: Adam weights 0.98 (huggingface default  0.999)
    lr_scheduler_type="linear",
    evaluation_strategy="epoch",
    save_strategy="steps",
    save_steps=50, # Checkpoint saves every 50 steps / every 100,000 reviews ( 50 * 10 * 34 * 6)
    save_total_limit=1, # Saves latest checkpoint
    torch_compile=True,  # Much Faster
    push_to_hub=True,
    hub_strategy="checkpoint", # Only pushes at end with save_model()
    logging_steps=25, # Logs training progress every 25 steps
)
```

```python
from transformers import DataCollatorForLanguageModeling

data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm_probability=0.15
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["validation"],
    data_collator=data_collator,
)
print_gpu_utilization()
```

```python
eval_results = trainer.evaluate()
non_trained_eval_loss = eval_results["eval_loss"]
print_gpu_utilization()
print(f"Non Trained Eval Loss: {non_trained_eval_loss:.2f}")
print(eval_results)
```

```python
start_time = time()
results = trainer.train()
training_time = time() - start_time
print(f"Training time: {training_time:.2f} seconds")
print_gpu_utilization()
print(results)
```

```python
eval_results = trainer.evaluate()
trained_eval_loss = eval_results["eval_loss"]
print_gpu_utilization()
print(f"Trained Eval Loss: {trained_eval_loss:.2f}")
print(eval_results)
```

```python
trainer.save_model()
```
